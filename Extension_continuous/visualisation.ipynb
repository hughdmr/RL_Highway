```python
import gymnasium as gym
import highway_env
import imageio
import numpy as np
import random
from pathlib import Path

NAME = '500k-1653-ls'
CKPT_DIR = Path(f"checkpoints/{NAME}")

def record_episode(model, env_config, path="episode.gif", seed=42):
    env = gym.make("highway-v0", render_mode="rgb_array")
    env.unwrapped.configure(env_config)

    obs, info = env.reset(seed=seed)
    frames = []
    done = False
    step = 0

    while not done:
        frame = env.render()
        if frame is None:
            raise RuntimeError("env.render() returned None. Check render_mode='rgb_array'.")
        frames.append(np.asarray(frame))

        speed = float(env.unwrapped.vehicle.speed)
        print(f"step={step} | speed={speed:.2f} m/s | {speed * 3.6:.1f} km/h")

        action, _ = model.predict(obs, deterministic=True)
        obs, reward, terminated, truncated, info = env.step(action)
        done = terminated or truncated
        step += 1

    env.close()
    imageio.mimsave(path, frames, fps=15)
    print(f"Saved → {path}")

#sac_best = SAC.load(str(CKPT_DIR / "sac" / "sac_ckpt_200000_steps.zip"))
sac_best = SAC.load(str(CKPT_DIR / "sac_best" / "best_model.zip"))

record_episode(sac_best, VISU_CONFIG, path="sac_episode.gif", seed=105)
```
